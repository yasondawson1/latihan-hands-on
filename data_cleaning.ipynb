# 1. Import

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("zomato_east_restaurants.csv",error_bad_lines=False)

dataset.head()
```

```python
print(len(dataset))
```

```python
def check_distribution(attr_name):
    print (dataset[attr_name].value_counts())
```

# 2. Drop Unknown Column

```python
dataset.drop(dataset.columns[[0]], axis=1, inplace=True)
dataset.tail()
```

# 3. Replace data inconsistency ('Rating')

```python
dataset['rating'] = dataset['rating'].str.strip()
dataset['rating'] = dataset.rating.apply(str)
#check_distribution('rating')

# analyse rating
dataset['rating'] = dataset.rating.replace(to_replace = "NEW", value ="0")
dataset['rating'] = dataset.rating.replace(to_replace = "-", value ="0")
dataset['rating'] = dataset.rating.replace(to_replace = "BARU", value ="0")
dataset.tail()
#check_distribution('rating')
#dataset.dtypes
```

# 4. Outlier and normalization ('cost_for_two') 

```python
#check_distribution('cost_for_two')
dataset['cost_for_two'] = dataset['cost_for_two'].replace(to_replace = '100', value ='100000')
dataset['cost_for_two']  = dataset.cost_for_two.map(lambda x: x.lstrip('p'))
dataset['cost_for_two'] = dataset['cost_for_two'].str.translate({ord('.'): None})
check_distribution('cost_for_two')

dataset.dtypes
```

# 5. Data Type Standarization

```python
dataset['cost_for_two'] = dataset['cost_for_two'].apply(int)
dataset['rating'] = dataset['rating'].apply(float)
dataset.dtypes
```

```python
dataset.head()
```

```python
dataset.shape
```

# 6. Remove (') , space from cuisines

```python
dataset['cuisines'] = dataset['cuisines'].str.strip()
dataset['cuisines'] = dataset['cuisines'].str.replace("'", '')
dataset['cuisines'] = dataset['cuisines'].str.replace("Asian", 'Asia')
dataset['cuisines'] = dataset['cuisines'].str.replace("Barbekyu", 'BBQ')
dataset['cuisines'] = dataset['cuisines'].str.replace("Barat", 'Western')
dataset['cuisines'] = dataset['cuisines'].str.replace("Kopi", 'Coffee')
dataset['cuisines'] = dataset['cuisines'].str.replace("Makanan Cepat Saji", 'Fast Food')
dataset['cuisines'] = dataset['cuisines'].str.replace("Indonesian", 'Indonesia')
dataset['cuisines'] = dataset['cuisines'].str.replace("Italian", 'Itali')
dataset['cuisines'] = dataset['cuisines'].str.replace("Japanese", 'Jepang')
dataset['cuisines'] = dataset['cuisines'].str.replace("Makanan Jalanan", 'Street Food')
dataset['cuisines'] = dataset['cuisines'].str.replace("Drinks Only", 'Minuman')
dataset['cuisines'] = dataset.cuisines.apply(lambda x: x[1:-1].split(','))
```

```python
cuisine_list = []
for i in range(0, 1675):
    cuisine_list.append(dataset.iloc[i]['cuisines'])
print(cuisine_list)
```

# 7. Unique Cuisines

```python
cuisine_unique_list = []
for i in range(0, 1675):
    cuisine_unique_list += cuisine_list[i]
print(cuisine_unique_list)
```

```python
cuisine_unique = set(cuisine_unique_list)
sorted_cuisine = sorted(cuisine_unique)
print(sorted_cuisine)
```

```python
sorted_cuisine = [s.strip() for s in sorted_cuisine]
sorted_cuisine = [s.replace("'", '') for s in sorted_cuisine]
print(len(sorted_cuisine))
```
